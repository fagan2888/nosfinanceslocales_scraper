```python
from localfinance.account_network import make_city_account
import networkx as nx
```

```python
city_account = make_city_account()
g = nx.DiGraph()
```

```python
for node, node_attr in city_account.nodes.items():
    g.add_node(node, node_attr)
```

```python
for node, successors in city_account.succ.items():
    for successor, successor_attr in successors.items():
        g.add_edge(node, successor)
```

```python
pos=nx.graphviz_layout(g,prog='dot')
nx.draw(g,pos)
```
